```python
country_list = list(set(age_gender_data['country_destination']))
```

```python
for country in country_list:
    country_data = age_gender_data[age_gender_data['country_destination']== country]
    for gender_type in ('male', 'female'):
        gender_country = country_data[country_data['gender']== gender_type]
        gender_country_plot = sns.barplot(sorted(gender_country["age_bucket"]), 
                                          gender_country["population_in_thousands"])
        gender_country_plot.figure.set_size_inches(11,5.5)
        gender_country_plot.figure.suptitle(gender_type.title() + 's Traveling to {}'.format(country), size = 20)
        plt.show()
```
